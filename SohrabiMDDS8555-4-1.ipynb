```python
import matplotlib.pyplot as plt
import numpy as np


# Define the function piecewise
def y_hat(x):
    y = np.zeros_like(x)
    for i, val in enumerate(x):
        if val < 1:
            y[i] = 1 + val
        else:
            y[i] = 1 + val - 2 * (val - 1) ** 2
    return y


# Create x values from -2 to 2
x_vals = np.linspace(-2, 2, 400)
y_vals = y_hat(x_vals)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$\hat{Y}$', color='blue')

# Mark the discontinuity at X = 1
x_jump_left = 1 - 1e-6
x_jump_right = 1 + 1e-6
y_jump_left = 1 + x_jump_left
y_jump_right = 1 + x_jump_right - 2 * (x_jump_right - 1) ** 2
plt.plot([1], [y_jump_left], 'ro')
plt.plot([1], [y_jump_right], 'go')

# Axis labels and legend
plt.axvline(x=1, color='gray', linestyle='--', linewidth=1)
#plt.title(r'Estimated Curve $\hat{Y} = 1 + X - 2(X-1)^2 I(X \geq 1)$')
plt.xlabel('X')
plt.ylabel(r'$\hat{Y}$')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
